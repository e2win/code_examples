```python
import numpy as np
import pandas as pd
import folium
from shapely.geometry import Polygon
import geopandas as gpd
import json
import matplotlib.pyplot as plt
import plotly.express as px
from dash import Dash, dcc, html, Input, Output
import plotly.graph_objs as go
import plotly.figure_factory as ff
from google.colab import output
output.enable_custom_widget_manager()
```

# Part 1

```python
df = pd.read_csv(google_file_prefix+"capitals_lat_lon_csv_2.csv")
print(df.head())
```

```python
map = folium.Map(location = [43.001525,41.023415], tiles='OpenStreetMap', zoom_start=6)
for i, row in df.iterrows():
    lat = df.at[i, 'Latitude']
    lng = df.at[i, 'Longitude']
    popup = 'Capital : ' + str(df.at[i, 'Capital']) + '<br>' + '<br>' +'Country: ' + df.at[i, 'Country']
    folium.Marker(location = [lat, lng], popup= popup, icon = folium.Icon(color='blue')).add_to(map)

map
```

```python
map.save(google_file_prefix+"World_Capital_Locations.html")
```

# Part 2

```python
Africa = {"Country": ["South Africa", 'Somalia', 'Morocco'], "Capital": ["Capetown", "Mogadishu", "Rabat"], "Latitude": [-33.918861, 2.046934, 33.97159], "Longitude":[18.423300, 45.318162, -6.849813]}
df_africa = pd.DataFrame(Africa)
print(df_africa)
```

```python
map_a = folium.Map(location = [33.97159,-6.849813], tiles = 'OpenStreetMap',                               
          zoom_start=6)
for i, row in df_africa.iterrows():
    lat = df_africa.at[i, 'Latitude']
    lng = df_africa.at[i, 'Longitude']
    popup = 'Capital : ' + str(df_africa.at[i, 'Capital']) + '<br>' + '<br>' +'Country: ' + df_africa.at[i, 'Country']
    folium.Marker(location = [lat, lng], popup= popup, icon = folium.Icon(color='blue')).add_to(map_a)

polygon = Polygon([(-6.849813, 33.97159),(18.423300,-33.918861),(45.318162,2.046934)])
folium.GeoJson(polygon).add_to(map_a)
map_a
```

```python
print("The area of the above polygon is:")
print(polygon.area)

print("The perimeter of the above polygon is:")
print(polygon.length)
```

```python

# based on stack-overflow answer
exp_array = np.random.exponential(scale = 3.2, size = 500)
fig = px.histogram(exp_array, nbins=15)
fig.show()

hist_data = [exp_array]
group_labels = ['distplot']
f = ff.create_distplot(hist_data, group_labels, bin_size=5, show_rug=False)
fig = go.FigureWidget(f)
fig.data[0].marker.line =  dict( color = 'black',width = 2)
fig.data[1].line.color = 'red'
fig.layout.sliders = [dict(
                active = 4,
                currentvalue = {"prefix": "bin size: "},
                pad = {"t": 20},
                steps = [dict(label = i, method = 'restyle',  args = ['xbins.size', i]) for i in range(1,20)]
            )]
fig
```
